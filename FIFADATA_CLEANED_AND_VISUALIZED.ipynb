```python
#visualization 
position_by_order = ["Strength", "Acceleration", "Sprint Speed","W/F" ,"SM", "A/W", "D/W", "IR", "PAC", "SHO", "PAS","DRI", "DEF" ,"PHY"]


position_colors = {"Strength": "blue", "Defending": "green", "W/F": "green","A/W" : "green","Acceleration": "green","PHY": "green","D/W": "green",
    "Sprint Speed": "orange","CM": "orange","RM": "orange","CAM": "orange","Attacking": "red","Dribbling": "red", "Penalties": "red", "ST": "red"
}
```

```python
# Select columns based on position_by_order list
selected_columns = datafifa[position_by_order]

# Plot density for each column
selected_columns.plot(kind="density", figsize=(12, 8))
plt.title("Density Plot for Positions")
plt.xlabel("Density")
plt.ylabel("Frequency")
plt.show()
```

```python
#side by side Boxplot comparing  Age and skill 
datafifa.boxplot(column="Age",
                by="Skill",
                color="blue",
                figsize=(8,8))
```

```python
#comparing columns
fifadata_table=pd.crosstab(index=datafifa["Age"], 
                           columns=datafifa["Dribbling"])
```

```python
fifadata_table.plot(kind="bar",
                   figsize=(8,8),
                   stacked=True)
```

```python
#To get better view 
fifadata_table.plot(kind="bar",
                   figsize=(8,8),
                   stacked=False)
```

```python
#Correlation Heatmap

df_s = datafifa[["Age","POT","Height","Weight","BOV","Value","Wage","Release Clause","A/W","D/W", "IR","PAC", "SHO", "PAS", "DRI" ,"DEF","PHY"]]

corr_matrix =df_s.corr()


mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(12, 8))

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", mask=mask)


plt.show()
```
